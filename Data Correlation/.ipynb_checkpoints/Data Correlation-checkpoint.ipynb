```python
import pandas as pd
import numpy as np
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv('data/movie_data.csv')
```

```python
df.head()
```

### Dealing with missing data 

```python
for col in df.columns:
    missingData = np.mean(df[col].isnull())
    print('{} = {}%'.format(col,missingData))
```

```python
df = df.fillna(method = 'ffill')
```

```python
for col in df.columns:
    missingData = np.mean(df[col].isnull())
    print('{} = {}%'.format(col,missingData))
```

```python
df.shape
```

### Changing types

```python
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
df['score'] = df['score'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
df.dtypes
```

```python
df
```

### Year Extraction

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int) #string patterns (4)
```

```python
df.head()
```

### Sorting

```python
df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
df.sort_values(by=['gross'], inplace = False, ascending = False)
```

### Show Correlation 

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title ('Budget vs Gross earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
plt.scatter(x=df['company'], y=df['gross'])
plt.title ('Company vs Gross earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Company')
plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
```

```python
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
```

```python
sns.heatmap(correlation_matrix, annot = True)
plt.show()
```

### String to number representation 

```python
numdf = df

for col_name in numdf.columns:
    if (numdf[col_name].dtype == 'object'):
        numdf[col_name] = numdf[col_name].astype('category')
        numdf[col_name] = numdf[col_name].cat.codes
    
numdf
```

```python
correlation_matrix = numdf.corr(method='pearson')
```

```python
sns.heatmap(correlation_matrix, annot = True)
plt.show()
```

```python
numdf.corr()
```

```python
correlation_mat = numdf.corr()

corr_pairs = correlation_mat.unstack()
sorted_corr = corr_pairs.sort_values()
sorted_corr
```

```python
high_corr = sorted_corr[(sorted_corr) > 0.5]
```

```python
high_corr
```
